Payment Method:

Frequency of Purchases:

Elegi estas variables porque se puede observar los habitos de compra por edades, tipo de prendas que se consume y publico arientado. Con esto se puede generar una estrategia de marketing eficiente y personalizada.

DATASET 2 - Mental Health Of Remote Workers
Este Dataset nos muestra informacion de empleados que trabajan bajo modalidad de Trabajo Remoto y algunas de las variables que podemos destacar son:

Age: El cual posee un promedio desde los 28 a los 49 años.Lo que nos indica que son trabajos para jovenes o de media edad.

Work_Mode: El modo de trabajo puede ser Hybrido, Remoto y Onsite. Esto nos muestra que el trabajo posee flexibilidad.

Mental_Health_Status y Productivity_Score: Moderado (Moderate), malo (poor) y Bueno (good).
Los que poseen un Status Mental Good poseen mejor puntuacion en la productividad, mientras que los Poor presentan una productividad mas baja.

Team_Communication_Effectiveness: La puntuacion de comunicacion en el equipo dentro de todo es muy buena, lo que puede significar que hay buena colaboracion independientemente de la modalidad de trabajo.

Elegi este segundo Dataset ya que podemos determinar la tendencia de como influye en la salud mental de los trabajadores la preferencia de como trabajar (hibrida, remota o presencial), dando asi a la empresa la posibilidad de entender lo importante que es mantener estas modalidades de trabajo destacando la flexibilidad laboral vs la vida personal.Tambien ayuda a identificar problemas en la empresa como la conectividad y poder sugerir acciones que la mejoren.
